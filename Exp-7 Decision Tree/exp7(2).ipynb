```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
import matplotlib.pyplot as plt
```

```python
data = {
    'weather': ['sunny', 'sunny', 'sunny', 'rainy', 'rainy', 'sunny', 'rainy', 'sunny'],
    'activity': ['go_out', 'go_out', 'go_out', 'stay_in', 'stay_in', 'go_out', 'stay_in', 'go_out']
}
```

```python
df = pd.DataFrame(data)
```

```python
df['weather'] = df['weather'].map({'sunny': 1, 'rainy': 0})
df['activity'] = df['activity'].map({'go_out': 1, 'stay_in': 0})
```

```python
X = df[['weather']]
y = df['activity']
```

```python
clf = DecisionTreeClassifier(criterion='entropy', max_depth=2, random_state=42)
clf.fit(X, y)
```

```python
tree_rules = export_text(clf, feature_names=['weather'])
print(tree_rules)
```

```python
plt.figure(figsize=(6,4))
plot_tree(clf, feature_names=['weather'], class_names=['stay_in', 'go_out'], filled=True)
plt.show()
```
